```python
# Chi square test = To check the dependancy of the variables, Variables should be categorical
```

```python
dataset3 =pd.read_excel("5 Chi square Test.xlsx",sheet_name =0)
```

```python
dataset3
```

```python
# now remove the null vakues
dataset4 = dataset3.dropna()
```

```python
dataset4
```

```python
from scipy.stats import chi2_contingency
```

```python
chitable = pd.crosstab(dataset4.Gender, dataset4.Smoking)
```

```python
chitable
```

```python
stats,p,dof,expected = chi2_contingency(chitable)
```

```python
print(stats,p)
```
