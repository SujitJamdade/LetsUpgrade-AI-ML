## skewness

skewness is use to give the symmetricness of data

s<0 Positive skewess : mean is more than the medium

S>0 Negative skewness : mean is less than the medium

s=0 no skewness : here mean is equal to the medium

```python
dataset[['CurrentSalary', 'After6Months', 'SalBegin']].skew()
```

```python
dataset[['CurrentSalary', 'After6Months', 'SalBegin']].mean()
```

```python
dataset[['CurrentSalary', 'After6Months', 'SalBegin']].median()
```

```python
# See in above 3 cell mean is more than the median so skewness is positive
```

## kurtosis
kurtosis is use to measure the peakness of the data

There are three types of kurtosis

leptokurtic : positive value

mesokurtic : zero value

platykurtic : negative value

```python
dataset[['CurrentSalary', 'After6Months', 'SalBegin']].kurt()
```
